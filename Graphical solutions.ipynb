```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

# 1. Maximizing Profit for a Factory 

# Define the constraints
x1 = np.linspace(0, 10, 400)

# Constraint 1: 2x1 + 3x2 <= 12 --> x2 <= (12 - 2x1)/3
x2_1 = (12 - 2 * x1) / 3

# Constraint 2: x1 + 2x2 <= 8 --> x2 <= (8 - x1) / 2
x2_2 = (8 - x1) / 2

# Non-negativity
x2_1[x2_1 < 0] = 0
x2_2[x2_2 < 0] = 0

# Plot the constraints
plt.figure(figsize=(8, 6))
plt.plot(x1, x2_1, label="2x1 + 3x2 <= 12")
plt.plot(x1, x2_2, label="x1 + 2x2 <= 8")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.fill_between(x1, np.minimum(x2_1, x2_2), color='gray', alpha=0.3, label='Feasible Region')
plt.xlim(0, 10)
plt.ylim(0, 5)
plt.xlabel("x1 (Product A)")
plt.ylabel("x2 (Product B)")
plt.title("Feasible Region for Maximizing Profit")
plt.legend()
plt.grid()
plt.show()

# Corner points of the feasible region
corners = [
    (0, 0),  # Origin
    (0, 4),  # Intersection with x2-axis
    (3, 2),  # Intersection of the two constraints
    (6, 0),  # Intersection with x1-axis
]

# Calculate objective function at corner points
profits = [3*x[0] + 4*x[1] for x in corners]

# Find the optimal solution
optimal_index = np.argmax(profits)
optimal_point = corners[optimal_index]
optimal_profit = profits[optimal_index]

print("Corner Points:", corners)
print("Objective Function Values:", profits)
print(f"Optimal Solution: x1 = {optimal_point[0]}, x2 = {optimal_point[1]}, Maximum Profit = {optimal_profit}")
```
